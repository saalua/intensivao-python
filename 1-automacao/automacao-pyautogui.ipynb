- pip install pyautogui - permite automatizar os movimentos do mouse e do teclado

```python
import pyautogui
import time
import pyperclip

# executa um comando e dá uma pausa de 1 segundo
pyautogui.PAUSE = 1 

# Passo 1
# Abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

# Entrar no link do sistema
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
# pyautogui.write('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga')

# Passo 2
# Para clicar precisamos passar a posição do eixo x e o eixo y, o terceiro parâmetro é a quantidade de cliques 
time.sleep(3)
pyautogui.click(1763, 311, clicks=2)

# Passo 3
pyautogui.click(1783, 354, button='right') #clicar no arquivo com o botão direito
#pyautogui.click(2530, 185) #clicar nos três pontinhos
pyautogui.click(1850, 645) #clicar na opção fazer download
```

```python
# Passo 4
# Ler o arquivo baixado
import pandas as pd

#aquele r no início mostra que não tem nenhum caractere especial do python, mostra que p texto é exatamente o que eu digitei
tabela = pd.read_excel(r'C:\Users\dsnpack\Downloads\Vendas - Dez.xlsx')  #sheets=2 caso seja outra aba
faturamento = tabela['Valor Final'].sum() #soma da coluna valor final
quantidade = tabela['Quantidade'].sum() #soma da coluna quantidade
display(tabela)
```

```python
# Passo 5
pyautogui.hotkey('ctrl', 't')
link = 'https://outlook.live.com/mail/0/inbox' #já estou logada
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(5)

pyautogui.click(1519, 172) #clicar no campo Nova mensagem
pyautogui.click(1808, 291)
pyautogui.write('EMAIL AQUI')
time.sleep(2)
pyautogui.press('enter')
pyautogui.press('tab') #para passar para o campo assunto
assunto = 'Relatório de vendas'
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

#pra escrever texto com mais de uma linha coloca-se entre 3 aspas
#esse f diz para o python que vai ter uma variável no meu texto
texto_email = f""" 
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Luana
"""

pyautogui.write(texto_email)

#clicar em enviar
pyautogui.hotkey('ctrl', 'enter')




```
